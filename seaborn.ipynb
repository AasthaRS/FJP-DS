```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

#### here 'tip' is dependant and rest are independant attributes

```python
df
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df)
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, size = 'size')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, size = 'day')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, hue = 'sex', size = 'size')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, hue = 'sex', style = 'time', size = 'size')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, hue = 'sex', kind='line', style = 'time', size = 'size')
```
